# Pandas Methods for Data Cleaning, indexing and Grouping

```python
# set up environment
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline

print('ready')
```

```python
url = "https://raw.githubusercontent.com/realpython/python-data-cleaning/master/Datasets/olympics.csv"

data = pd.read_csv(url, header=1)

data.head(3)
```

```python
df = data.copy() # deep copy 
```

```python
df.head(2)
```

```python
# column names 
df.columns
```

```python
df.columns = df.columns.str.replace("?",'').str.replace("!",'').str.replace("0",'').str.replace(" ",'')
```

```python
df.head(2)
```

```python
# dataframe rename()

df = df.rename(columns={
    df.columns[0] : 'country',
    'Summer': 'summergames',
    df.columns[2] : 'summer_G',
    df.columns[3] : 'summer_S',
    df.columns[4] : 'summer_B',
    'Total': 'summermedals',
    'Winter': 'wintergames',
    df.columns[7] : 'winter_G',
    df.columns[8] : 'winter_S',
    df.columns[9] : 'winter_B',
    'Total.1': 'wintermedals',
    'Games': 'totalgames',
    df.columns[12] : 'total_gold',
    df.columns[13] : 'total_silver',
    df.columns[14] : 'total_bronze',
    df.columns[15] : 'totalmedals(S+W)'
})
```

```python
# df.to_string()
pd.set_option("max_rows", None)
```

```python
df.head(3)
```

```python
# split by delimiters
df["country"].str.split("(")[0]
```

```python
temp = df["country"].str.split("(", n=1, expand=True)
```

```python
temp.head(3)
```

```python
df.country = temp[0]
```

```python
df.head(2)
```

```python
data.head(2)
```

```python
df.tail(3)
```

```python
### save the file after cleaning

df.tail(2)

# what file type csv, excel

# df.to_csv('filename' + filepath)

df.to_csv('./datasets/olympics_cleaned.csv', index=False)
```
