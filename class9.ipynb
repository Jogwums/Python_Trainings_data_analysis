# Visualization in Python

### Matplot lib and Seaborn 

```python
# set up your environment 
import numpy as np 
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns 

print('done')
```

```python
url = './datasets/survey.csv'

df = pd.read_csv(url)

df.head(3)
```

```python
plt.plot([1,2,3,4],[2,4,6,8]) # global api
```

```python
# features are added using dot notation, within the same block
# title, axis labels (x axis label, y axis label) etc.

x = [1,2,3,4]
y = [1,4,2,6]

plt.plot(x,y)
plt.title("Simple plot")
plt.xlabel("rise")
plt.ylabel("fall")
plt.show()
```

```python
# Object oriented API 
```

```python
fig, ax = plt.subplots()

ax.plot(x,y)
```

```python
fig = plt.figure()
fig, ax = plt.subplots()
fig, axes = plt.subplots(2,2)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('Simple plot')
axes[1].plot(y,x)
```

# bar charts

```python
count = len(df["Hair Color"] == "Black")
print(count)
df.dropna(inplace=True)
categories = df["Hair Color"].unique()

fig, ax = plt.subplots(figsize=(6,4), layout='constrained')

ax.bar(categories, count)
```

```python
movies = ['Star wars', "007", 'Finding Nemo', 
          "Tom and Jerry", "Fast and Furious"]
num_awards = [10, 8, 4, 5, 3]
```

```python
print(len(movies))
print(len(num_awards))
```

```python
# global api 
plt.figure(figsize=(12,6))
plt.bar(movies, num_awards, color='yellow')
```

```python
plt.figure(figsize=(12,6))
plt.bar(movies, num_awards, color=['red','blue','green'], width=0.4)
```

```python
plt.figure(figsize=(12,6))
plt.barh(movies, num_awards, color=['pink','black','red','blue','green'], height=0.4)
```

# Histogram

#### used to show frequency 
#### it shows the number of observations within each given interval

```python
# Example 
x = np.random.normal(170,10,250)
```

```python
# the code above generates 250 random values 
# representing the height of 250 people 
# the values center around 170, with a standard 
# deviation of 10 
```

```python
plt.hist(x)
plt.title("Height of 250 respondents")
plt.xlabel("height (cm)")
plt.ylabel('no. of people')
plt.show()
```

```python
# using subplots
plt.figure(figsize=(12,6))
plt.title("subplots")
#
plt.subplot(1,2,1) # 1 row, 2 columns, 1st panel 
plt.plot(x)

plt.subplot(1,2,2) # 1 row, 2 columns, 2nd panel
plt.hist(x, color='green')

plt.show()
```

```python
# crude oil data 

url = './datasets/crudeoil.csv'

crude = pd.read_csv(url, header=0)

del crude["Unnamed: 5"]

crude.head(3)
```

```python
prod = crude.Production
export = crude["Crude Oil Export"]
```

```python
# plt.scatter(prod, export)
```

```python
year = crude["Year"]

plt.figure(figsize=(12,6))
plt.scatter(year, prod)
plt.title("Oil Production - Nigeria 2006 to 2021")
plt.xlabel("Year")
plt.ylabel("Oil Production (mpd)")
plt.show()
```

```python
# line chart
```

```python
plt.figure(figsize=(12,6))
plt.plot(year, prod)
plt.title("Oil Production - Nigeria 2006 to 2021")
plt.xlabel("Year")
plt.ylabel("Oil Production (mpd)")
plt.show()
```

```python
# seaborn boxplot
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=year,y=prod,data=crude)
sns.set_style('darkgrid')
plt.xticks(rotation=90)
plt.show()
```
