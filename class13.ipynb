# Applying functions to groupby groups

we can apply functions to each group by using the pandas apply() method

```python
def get_agg(group):
    return {'min': group.min(),
            'max': group.max(),
            'avg': group.mean(),
            'total': group.sum()}
```

```python
print(grp.apply(get_agg))
```

```python
get_agg(grp)
```
