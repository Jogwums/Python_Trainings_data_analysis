# Pandas Groupby 

```python
survey = pd.read_csv('./datasets/survey.csv')
olympics = pd.read_csv('./datasets/olympics_cleaned.csv')
```

```python
olympics.head(3)
```

```python
# Pandas group by works in 3 steps 
# 1. split the table into groups 
# 2. apply some operations to each of the smaller 
#    tables 
# 3. combine the results 
```

```python
country_grp = olympics.groupby('country')
```

```python
country_grp.first()
```

```python
country_grp.groups
```

```python
country_sur = survey.groupby('Country')
```

```python
country_sur.groups
```

```python
survey2 = survey.copy()
```

```python
survey2.columns
```

```python
survey2['Country'] = survey2['Country'].str.replace('NIGERIA','Nigeria').str.replace('Nigeria ','Nigeria').str.replace('Nigerian','Nigeria').str.replace('nigeria','Nigeria').str.replace('American ','America')
```

```python
survey2.Country.unique()
```

```python
survey2
```

```python
survey2.loc[11,"Country"] = 'Nigeria'
```

```python
survey2.Country
```

```python
survey2.Country.unique()
```

```python
# Plot a chart
```

```python
plt.figure(figsize=(12,5))
nums_c = survey2.Country.value_counts()
# print(nums)
nums_c.plot(kind='bar')
plt.title('Students by Country')
plt.ylabel('students')
plt.xlabel('country')
plt.show()
```

```python
survey2.columns
```

```python
# gender counts
plt.figure(figsize=(12,5))
nums_g = survey2.Gender.value_counts()
nums_g.plot(kind='bar')
plt.title('Students by Gender')
plt.ylabel('students')
plt.xlabel('gender')
plt.show()
```

```python
plt.figure(figsize=(12,5))
nums_h = survey2["Hair Color"].value_counts()
nums_h.plot(kind='bar')
plt.title('Students Hair Color')
plt.ylabel('students')
plt.xlabel('hair color')
plt.show()
```
